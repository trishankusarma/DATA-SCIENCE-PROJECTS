```python
import numpy as np
```

```python
def getNRandomInt(n,N):
    A = np.random.randint(N, size=(n));
    return A;
    
X=getNRandomInt(100,10)

print("Random array of 100 numbers with random elements within the range of 0 to 10\n")

print(X)
```

```python
min = np.min(X)
print("Minimum value:\n",min)
```

```python
max = np.max(X)
print("Maximum value:\n",max)
```

```python
avg = np.mean(X)
print("The average value is :\n",avg)
```

```python
unique_elements , counts_elements = np.unique( X , return_counts = True)

print("Frequency of unique values of the said array is :\n")

for x in unique_elements :
    print(" Element ", x ," : Frquency ", counts_elements[x])
```

```python
from matplotlib import pyplot as plt

#plotting to our canvas
plt.bar(unique_elements , counts_elements )

plt.title("Frequency distribution")
plt.xlabel("Numbers")
plt.ylabel("Frequency of the numbers")

#frequency distribution table
plt.show()
```

**Answer to question 7**

```python
import numpy as  np
import math
from matplotlib import pyplot as plt

def plotValue(c,k):
    powV = np.power(c,k);
    expV = np.exp(-c);
    factV = math.factorial(k);
    return (powV*expV)/factV; 
```

```python
kArray = np.arange(51);
functionArray = np.empty(51);

for x in kArray :
    functionArray[x] = plotValue(1,x);

plt.plot( kArray , functionArray );

plt.title("Function F(k)=(C^k * e^-C)/(k!) Where C is a constant, k is a positive integer, e is eulerâ€™s constant. C : 1 ");
plt.xlabel("k - value");
plt.ylabel("function value for corresponding value of k");

plt.show();
```

**Answer to question 8**

```python
import numpy as np;
import math;
from matplotlib import pyplot as plt;

position=0;
i=0;

#Array with values from -100 to 100
arr = np.linspace(-100,100,201);

#create an array to store the value of the occurence of a particular position
freq = np.zeros(201, dtype = int);
#-100 to 100 

freq[100]=1;
#As 0 is the initial position

print("For the sake of understanding we assume on a random trail, x=0 meaning a tail and so position shift to left by 1 and x=1 means head and position shift to right by 1\n")

while i<1000:
    x=math.floor(np.random.rand()*2)
    if x==0:
        position-=1;
    else:
        position+=1;
    freq[position+100]+=1
    i+=1

plt.plot(arr,freq);

plt.title("e frequencies against each cell value using matplotlib")
plt.xlabel("cell values")
plt.ylabel("Frequencies of occurence of each cell")

plt.show()
```

```python
#!/usr/bin/env python
# coding: utf-8

# In[ ]:


import numpy as np;
import math;
from matplotlib import pyplot as plt;

position=0;
i=0;

#Array with values from -100 to 100
arr = np.linspace(-100,100,201);

#create an array to store the value of the occurence of a particular position
freq = np.zeros(201, dtype = int);
#-100 to 100 

freq[100]=1;
#As 0 is the initial position

print("For the sake of understanding we assume on a random trail, x=0 meaning a tail and so position shift to left by 1 and x=1 means head and position shift to right by 1\n")

while i<1000:
    import random
    coin= ['tail','head']
    x=random.choice(coin)
    if x=='tail':
        position-=1;
    else:
        position+=1;
    freq[position+100]+=1
    i+=1

plt.plot(arr,freq);
plt.xlabel("cell values")
plt.ylabel("Frequencies of occurence of each cell")

plt.show()
```

**Answer to question 3**

WAP to declare a blank list. Using loop (for/while your choice) append 10 random
numbers to the list. Display the content of the list.



```python
# creating a blank list
import random;
import math;

list = [];

#appending 10 random numbers to the list
i=0;
while i<10 :
    n = math.floor(random.random()*100);
    list.append(n);
    i+=1;

#display contents of list

for x in list:
    print(x)
```
